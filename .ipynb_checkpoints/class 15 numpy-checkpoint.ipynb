### np.sort
- Return a sorted coopy of an array

https://numpy.org/doc/stable/reference/generated/numpy.sort.html

```python
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)[::-1]
```

```python
np.sort(b,axis = 0) # each column is sorted 
```

### np.append
- The numpy.append() appends value along the mentioned axis at the end of the array

https://numpy.org/doc/stable/reference/generated/numpy.append.html

```python
np.append(a,200)
```

```python
b
```

```python
np.append(b, np.random.random((b.shape[0],1)) , axis=1 )
```

- **b:** This is your original $6 \times 4$ integer array.b.shape[0]: This grabs the number of rows in b, which is 6.
- **np.random.random((6, 1)):** This generates a new 2D array filled with random decimals between 0 and 1. By passing (6, 1) as the shape, you ensure it has 6 rows and 1 column, matching the vertical height of array b.
- **axis=1:** This tells NumPy to append the new data horizontally (column-wise). If you don't specify an axis, NumPy will flatten both arrays into a single 1D line of numbers.

### np.concatenate()
- numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d) , axis=0)
```

### np.where
- The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
a
```

```python
# find all indices with value greater than 50
np.where(a>50)
```

```python
# replace all values > 50 with 0
np.where(a>50,0,a)
```

```python
np.where(a%2 == 0,0,a)
```

### np.corrcoef
- Return Pearson product-moment correlation coefficients.

https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html

```python
salary = np.array([20000,40000,25000,35000,60000])
experience = np.array([1,3,2,4,2])

np.corrcoef(salary,experience)
```

### Set functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(n,m)
```

```python
np.setxor1d(m,n)
```

```python
m[np.in1d(m,1)]
```
