## Write OOP classes to handle the following scenarios
- A user can create a view 2D co-ordinates.
- A user can find out the distance between 2 co-ordinates.
- A user can find the distance of co-ordinate from origin.
- A user can check if a point lies on given line.
- A user can find the distance between a given 2D point and a given line 

````python
class Point:
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cod , self.y_cod)

    def euclidean_dist(self,other):
        return ( (self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2 ) ** 0.5

    def distance_from_origin(self):
        return self.euclidean_dist(Point(0,0))

class Line: # Equation of line is :- Ax + By + C
    def __init__(self,A,B,C):
        self.a = A
        self.b = B
        self.c = C

    def __str__(self):
        return f'{self.a}x + {self.b}y + {self.c}'
        
    def point_lies_on_line(self,point):
        flag = ((self.a * point.x_cod) + (self.b * point.y_cod) + self.c)
        if flag == 0:
            return True
        else:
            return False

    def distance_point_line(self,point):
        num = abs((self.a*point.x_cod) + (self.b*point.y_cod) + (self.c)) 
        den = ( (self.a)**2 + (self.b)**2 )**0.5
        return (num/den)
````

````python
p1 = Point(0,0)
p2 = Point(1,1)
print(p1)
print(p2)
print(f'euclidean distance between p1 and p2 is {p1.euclidean_dist(p2)}')
print(f'Distance from origin p2 {p2.distance_from_origin()}')

l1 = Line(3,4,5)
print(l1)
print(f'Point (1,-2) lies on Line 3x + 4y + 5 {l1.point_lies_on_line(Point(1,-2))}')
print(f'Distance between p2 and line {l1.distance_point_line(p2)}')
````

# Ecapsulation :- 



## 1. Encapsulation (The Capsule / Medication Analogy)

Think of a class like a medical **pill capsule**. Inside the capsule, all the medicine is mixed together, safely sealed away from the outside world.

**Encapsulation means hiding your data so nobody can accidentally break it.** You don't let outside code touch your variables directly. Instead, you put "security guards" (methods) in front of them to make sure any changes are safe.

### How we do it in simple code:

If you put a double underscore `__` in front of a variable name, Python locks it. It becomes **private**.

```python
class BankAccount:
    def __init__(self, cash):
        self.__balance = cash  # The '__' makes it PRIVATE. Locked away!

    # This is a safe window (Getter) to let people LOOK at the balance
    def get_balance(self):
        return self.__balance

    # This is a safe door (Setter) to let people CHANGE the balance safely
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Hey! You can't deposit negative money!")

account = BankAccount(100)
# account.__balance = -5000  # CRASH! Python blocks this direct change.
account.deposit(50)          # This is allowed because it goes through our safe door.

```


## Quick Summary Cheat Sheet:

1. **Encapsulation:** Put data in a safe capsule. Lock it with `__` so it doesn't get messed up.
2. **Inheritance:** Pass down traits from a Parent class to a Child class to save typing time.
3. **Polymorphism:** Use the exact same method name (like `make_sound`) on different objects, but let them do it their own way.
4. **Abstraction:** Hide the scary background engine logic; only show the user simple buttons to interact with.

````python
class BankAccount:
    def __init__(self, cash):
        self.__balance = cash  # The '__' makes it PRIVATE. Locked away!

    # This is a safe window (Getter) to let people LOOK at the balance
    def get_balance(self):
        return self.__balance

    # This is a safe door (Setter) to let people CHANGE the balance safely
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Hey! You can't deposit negative money!")

account = BankAccount(100)
account.__balance = -5000  # CRASH! Python blocks this direct change.
# print(account.__balance)
account.deposit(50)          # This is allowed because it goes through our safe door.
````

## Static Variable VS Instance Variable
The fundamental difference between a **Static variable** (also called a **Class variable**) and an **Instance variable** comes down to **who owns the data** and **how it is stored in your computer's memory**.

Here is the simple, direct breakdown.

---

## 1. Quick Definitions

* **Instance Variable:** Owned by a **specific object**. Every single object you create gets its own isolated copy of this variable. If you change it in object A, object B is completely unaffected.
* **Static / Class Variable:** Owned by the **Class itself**. There is only **one single copy** of this variable in memory, shared globally by every object created from that class. If one object changes it, everyone sees the change.

---

## 2. Side-by-Side Comparison

| Feature | Instance Variable | Static / Class Variable |
| --- | --- | --- |
| **Who owns it?** | The individual **Object (Instance)**. | The entire **Class** blueprint. |
| **Where is it defined?** | Inside the constructor (`def __init__(self):`) using **`self.`** | Directly inside the class body, **outside** of any methods. |
| **Copies in memory?** | Multiple copies (one per object). | Exactly **one copy** total. |
| **How do you access it?** | `object_name.variable` | `Class_Name.variable` |
| **What is it used for?** | Storing unique data specific to an object (e.g., a user's name, email, or balance). | Storing shared, constant, or counter data (e.g., counting total users created, company name, API URLs). |

---

## 3. Clear Code Example

Let's look at a `User` class to see how they look and behave side-by-side:

```python
class User:
    # 1. STATIC VARIABLE (Shared by everyone)
    total_users_count = 0  

    def __init__(self, username):
        # 2. INSTANCE VARIABLE (Unique to each user)
        self.username = username  
        
        # Every time a new user is made, we increase the shared global counter
        User.total_users_count += 1

# Let's create two separate user objects
user1 = User("Alice")
user2 = User("Bob")

# --- Testing Instance Variables ---
print(user1.username)  # Output: Alice
print(user2.username)  # Output: Bob (They are totally separate!)

# --- Testing Static Variables ---
print(user1.total_users_count)  # Output: 2
print(user2.total_users_count)  # Output: 2
print(User.total_users_count)   # Output: 2 (Best practice: Access directly via the Class)

```

---

## 4. The Common Trap (Don't shadow your variables!)

In Python, you *can* read a static variable using an object name (like `user1.total_users_count`), but trying to **change** it that way creates a subtle bug called shadowing.

```python
class Smartphone:
    brand = "Apple"  # Static variable

phone1 = Smartphone()
phone2 = Smartphone()

# TRAP: Trying to change a static variable using an instance
phone1.brand = "Samsung" 

print(phone1.brand)  # Output: Samsung (It looks like it worked, BUT...)
print(phone2.brand)  # Output: Apple (phone2 didn't change!)
print(Smartphone.brand) # Output: Apple

```

### What happened here?

When you ran `phone1.brand = "Samsung"`, Python didn't change the shared static variable. Instead, it got confused and secretly created a **brand-new instance variable** named `brand` *only* inside `phone1`.

> **The Rule of Thumb:** If you want to change a Static variable, **always** use the Class name: `Smartphone.brand = "Samsung"`. If you want to change an Instance variable, use the `self` or object pointer: `self.name = "New Name"`.
