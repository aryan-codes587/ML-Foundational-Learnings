## Core Fundamentals :- 
- Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around data, or objects, rather than functions and logic. In Python, everything is an object. To build custom objects, you must first design their blueprints using classes.

### Classes and Objects :-
***The Blueprint Analogy***
- A Class is an abstract blueprint, template, or structural layout definition. It defines what attributes and behaviors an entity will have, but holds no physical presence on its own.

- An Object (or Instance) is the physical manifestation built directly from that blueprint. It occupies a distinct space in system heap memory and contains real, live data.

````python
class Atm:

    # Constructor
    def __init__(self):
        self.pin = ''
        self.balance = 0
        # print(self)

    def menu(self):
        user_input = input("""
        Hi How can i Help you
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        elif user_input == '5':
            exit()
            
    def create_pin(self):
        user_pin = input('Enter your pin')
        self.pin = user.pin

        user_balance = int(input('Enter balance'))
        self.balance = user_balance

        print('pin created successfully')
        self.menu()

    def change_pin(self):
        entered_pin = input('Enter existing pin')
        if self.pin == entered_pin:
            self.pin = input('Enter new pin')
        else:
            print("Your pin doesn't match")
            self.menu()

    def withdraw(self):
        user_pin = input('Enter pin')
        if self.pin == user_pin:
            # allow to withdraw
            amount = int(input('Enter the amount'))
            if amount <= self.balance:
                self.balance -= amount
                print(f'Withdraw successfull , Balance is {self.balance}')
            else:
                print('Insufficient Amount in account')
        else:
            print('Please enter correct pin')
        self.menu()

    def check_balance(self):
        user_pin = input('Enter pin')
        if self.pin == user_pin:
            print(f'Current Balance is {self.balance}')
        else:
            print('Please enter correct pin')
        self.menu()
````

### Use of Constructors

* Normally, after an object is created, the user decides which methods to call and when to call them.
* For example, in an ATM class, the user may choose to create an account, change the PIN, check the balance, or withdraw money.
* A constructor is different because it is executed automatically whenever an object is created, regardless of the user's actions.
* Constructors are useful for initializing attributes that the application designer does not want users to control directly.
* In an ATM system, attributes such as the account PIN and balance can be initialized and managed through the constructor.
* This helps ensure data integrity and prevents users from manipulating critical information in unintended ways.
* By using constructors, the creator of the application retains control over essential object initialization and security-related data.

***Name of constructor cannot be changed***

## Concept of `self`:- 


Think of a class as a **photocopy blueprint** for building a house.

In Python, **`self` is that exact pointer. It means "this specific object instance right here."**

---


```python
class Dog:
    def __init__(self, name):
        self.name = name  # "Assign this name to THIS specific dog"

    def bark(self):
        print(f"{self.name} says Woof!")  # "Print the name of THIS specific dog"

```

Now, let's create two separate dog objects from that single blueprint:

```python
dog1 = Dog("Rex")
dog2 = Dog("Buddy")

```

When you call `dog1.bark()`, how does Python know to print "Rex" instead of "Buddy"?

Behind the scenes, Python secretly translates your code into this:

```python
# What you write:
dog1.bark()

# What Python actually executes:
Dog.bark(dog1)  # <-- It passes dog1 directly into the 'self' slot!

```

Because `dog1` was passed in, `self.name` instantly becomes `dog1.name`, which evaluates to **"Rex"**.

---

## Why do we have to type it out every single time?

In languages like Java or C++, this pointer is hidden automatically (using a secret word called `this`).

Python's guiding philosophy is **"Explicit is better than implicit."** Python creators wanted everything to be completely transparent. If a function needs to know which object it is working on, it must state it openly as the very first parameter: `def method_name(self):`.

### Summary Checklist:

* `self` is not a special Python keyword; it's just a variable name standard. You *could* name it `cookie`, but your code would look confusing to other programmers. Always use `self`.
* `self` acts as a placeholder for **whichever object is currently calling the method**.
* It allows your objects to keep their data isolated so they don't accidentally overwrite each other's variables in memory.

````python
class Fraction:
    # parametrized constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num , self.den)

    def __add__(self,other):
        new_num = (self.num * other.den) + (other.num * self.den)
        new_den = self.den * self.num
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,other):
        new_num = (self.num * other.den) - (other.num * self.den)
        new_den = self.den * self.num
        return '{}/{}'.format(new_num,new_den)

    def convert_to_decimal(self):
        return self.num/self.den
````

````python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
````

````python
fr1.convert_to_decimal()
````

````python
print(fr1 + fr2)
````
