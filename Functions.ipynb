## The def Keyword & Defining a Function
To create a new function, use the def keyword, followed by the function name in lowercase (standard snake_case convention), parentheses for parameters, and a colon.

Because Python uses whitespace to group code blocks, the entire body of your function must be indented by 4 spaces.

````python
def make_coffee():
    # The function body is indented
    print("Grinding beans...")
    print("Brewing coffee...")
````

### Calling Functions
Defining a function merely registers its instructions in memory; it does not execute them. To run the code inside your function, you must call (invoke) it by writing its name followed by parentheses.

````python
# Calling the function defined above
make_coffee()
````

### Let's create a function with docstring
A docstring (documentation string) is a triple-quoted literal string placed immediately below the function definition line. It serves as official documentation detailing what the function does, what inputs it expects, and what it returns.

````python
def is_even(num):
    """
    This function returns if a given number is odd or even
    input - any valid integer
    output - odd/even
    """
    if type(num) == int:
        if num % 2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return 'Enter an integer'
print('Program to check if a number is even or odd')
a = int(input('Enter a number'))
print(f'The given number is {is_even(a)}')
````

### Parameters VS Arguments
- Though often used interchangeble there is clear distinction between them
- **Parameters** :- They are variable placeholders written inside the function definition
- **Arguments** :- The actual data values you pass into those placeholders when you call the function

````python
def greet_user(username):  # 'username' is a PARAMETER
    print(f"Hello, {username}")

greet_user("Alice")       # "Alice" is an ARGUMENT
````

### The return Statement
The return statement immediately terminates a function's execution and sends a value back out to the line of code that called it.

````python
def add(a, b):
    return a + b  # Exits and sends back the sum

result = add(5, 10) # 'result' catches the returned value (15)
````

***⚠️ The Invisible Return: If a function finishes its execution without hitting an explicit return statement, Python automatically returns None behind the scenes.***

### Positional Arguments
By default, Python matches arguments to parameters based strictly on the order in which you pass them. If you mix up the order, the values map to the wrong variables.

````python
def display_profile(name, age):
    print(f"Name: {name}, Age: {age}")

# Sent positionally: "Bob" -> name, 25 -> age
display_profile("Bob", 25)
````

### Keyword Arguments
Keyword arguments allow you to pass values by explicitly naming the parameters during the function call. When using this syntax, the order of the arguments no longer matters.



````python
# Order is completely flipped, but names guarantee correct mapping
display_profile(age=32, name="Charlie")
````

### Default Parameters
You can provide fallback default values for parameters in your function definition. This makes those arguments entirely optional when calling the function.

````python
def powers(base , exponent=2):
    return base**exponent
print(powers(2))
print(powers(2,3))
````

**⚠️ The Syntax Constraint: Default parameters must always follow non-default parameters in your function declaration lines. Writing def test(a=1, b): causes a syntax error.**

### *args (Arbitrary Positional Arguments)
When you don't know ahead of time how many positional arguments a user might pass, prefix a parameter with a single asterisk ****args***. Python packs all incoming arguments into a single, clean tuple.

````python
def sum_all_nume(*args):
    print(args)
    return sum(args)
print(sum_all_nume(2,3,4,5,6,7,8,9))
````

### kwargs (Arbitrary Keyword Arguments)
To accept an unpredictable number of named keyword arguments, prefix a parameter with two asterisks kwargs. Python packs these arguments into a standard mutable dictionary, where keys represent parameter names.

````python
def save_user_profile(id, **kwargs):
    print(kwargs)  # Internal view: {'email': 'a@b.com', 'role': 'Admin'}

save_user_profile(id=501, email="a@b.com", role="Admin")
````

### Local VS GLobal Scope
- Scope dicatated where specific variable can be read or modified inside your application
    - **GLobal Scope :-** Variable defined outside the function block. They can be read anywhere in your script
    - **Local Scope:-** Variable created ***inside*** function. They exist only while that function is executing and are completely inaccessible from the outside world

````python
global_var = 'I am global'

def scope_demo():
    local_var = 'I am local'
    print(global_var) # Accessible
scope_demo()
# print(local_var) # Crashes with NameError
````

### The **GLobal** & **nonLocal** keywords
#### A. The global keyword
- If you need to ***modify*** a global variable from inside a function you must ***explicitely*** declare it using ***global*** keyword.

````python
counter = 0
def increment():
    global counter
    counter += 1
increment()
print(counter)
````

### The nonlocal Keyword
Used inside nested functions, nonlocal tells Python to modify a variable belonging to the immediate outer (enclosing) function's scope, rather than creating a new local variable.

````python
def outer_function():
    msg = 'Hello'

    def inner_function():
        nonlocal msg # Targets the present function's variable
        msg = 'Changes made!!!' 

    inner_function()
    print(msg)

outer_function()
````

### Lambda Functions (Anonymous Functions)
Lambda functions are **quick, inline, one-liner functions that don't need a name**. They are ***defined using the lambda keyword*** and automatically return the result of their single expression.

Formulae :- 

`labmda parameters: expression`

````python
square = lambda x: x**2
print(square(5))

pairs = [(1,'one') , (3,'three') , (2,'two')]
pairs.sort(key=lambda item:item[0]) # Sorts list based on first elements of tuple
````

````python
def loud_greet(text):
    return text.upper() + '!'

def run_operation(func_to_run , value):
    return func_to_run(value)

print(run_operation(loud_greet , 'welcome'))
````

````python
# HOF 
def square(x):
    return x**2

def cube(x):
    return x**3

def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
    return output

square_list = transform(square , [1,2,3,4,5,6,7])
cube_list = transform(cube , [1,2,3,4,5,6,7])
print(f'Square list is :- {square_list}')
print(f'Cube list is :- {cube_list}')
````

### Recursion
Recursion is an algorithmic technique where a function calls itself to break a complex problem down into manageable sub-problems. Every recursive function requires a base case to stop the execution loop; otherwise, it will recurse infinitely and crash with a RecursionError (Stack Overflow).

````python
def factorial(n):
    # 1. Base Case (Stops the infinite loop)
    if n == 1:
        return 1
    # 2. Recursive Case
    return n * factorial(n - 1)

print(factorial(5)) # Calculations: 5 * 4 * 3 * 2 * 1 = 120
````

### Q2: Explain the difference between Python's parameter passing mechanism vs. traditional "Pass-by-Value" or "Pass-by-Reference".

#### Answer:

Python uses a mechanism called **Pass-by-Assignment** (or *Pass-by-Object-Reference*).

* When you pass an argument to a function, Python passes the *object reference* (pointer) by value.
* If the object is **mutable** (like a list or dict), modifications inside the function alter the original object outside.
* If the object is **immutable** (like an integer, string, or tuple), reassigning it inside the function simply rebinds the local variable reference to a new object, leaving the external object unchanged.

### Q5: What is the LEGB rule in Python? Explain the lookup hierarchy.

#### Answer:

The **LEGB** rule dictates the order of namespaces Python searches through to resolve variable references:

1. **L (Local):** Names assigned within a function block and not declared global.
2. **E (Enclosing):** Names in the local scope of any and all enclosing (parent) nested functions, from inner to outer.
3. **G (Global):** Names assigned at the top-level of the module file or declared global inside a function.
4. **B (Built-in):** Preloaded names in Python’s built-in system module (e.g., `print`, `ValueError`).

### Q6: What is a closure in Python, and how does it relate to the `nonlocal` keyword?

#### Answer:

A **closure** is a nested function object that retains access to variables in its enclosing parent scope, even after the parent function has completely finished executing.

If the inner function needs to **modify** the string or integer variable in that retained outer scope (rather than just reading it), it must use the `nonlocal` keyword to prevent Python from creating a new local variable shadowing it.

```python
def dynamic_counter():
    count = 0
    def increment():
        nonlocal count  # Preserves closure state modification
        count += 1
        return count
    return increment

counter_func = dynamic_counter()
print(counter_func())  # 1
print(counter_func())  # 2

```

### Q7: What will the following execution output, and how do you fix it?

```python
functions = []
for i in range(3):
    functions.append(lambda: i)

print([f() for f in functions])

```

#### Answer:

* **Output:** `[2, 2, 2]`
* **Why:** Python's closures bind to the *variable* itself, not to its temporary value at creation time. By the time the lambdas are actually invoked in the list comprehension, the loop has completed, and the variable `i` inside the outer scope evaluates to `2`.
* **Fix:** Force early variable binding using a default parameter:
```python
functions = [lambda i=i: i for i in range(3)]
print([f() for f in functions])  # Output: [0, 1, 2]

```

### Q8: What is the difference between global variables and the `global` keyword? Why is using `global` extensively considered bad architectural practice?

#### Answer:

A global variable is read-accessible anywhere within a module. The `global` keyword is an explicit instruction used inside a function block to allow *writing/reassigning* a top-level global variable.

Extensive use of `global` is an anti-pattern because it introduces **hidden state side-effects**. It breaks predictability, makes unit testing incredibly difficult (since functions depend on external context), and prevents clean parallel/thread execution.

### Q11: How do you optimize a CPU-bound recursive function (like Fibonacci calculation) without changing the recursive algorithm structural framework?

#### Answer:

You can implement **memoization** using the built-in `functools.lru_cache` decorator. This caches calculated results in an internal lookup dictionary, shifting the time complexity profile from an exponential $O(2^n)$ down to a linear $O(n)$.

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci(n):
    if n < 2: return n
    return fibonacci(n-1) + fibonacci(n-2)

```

### Q14: Explain what a Decorator is. Write a basic structural custom decorator that times how long a function takes to execute.

#### Answer:

A **decorator** is a design pattern used to extend or modify the behavior of a function or method dynamically without changing its source code.

```python
import time
from functools import wraps

def execution_timer(func):
    @wraps(func)  # Preserves target function's metadata (__name__, __doc__)
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        print(f"{func.__name__} executed in: {end_time - start_time:.6f} seconds")
        return result
    return wrapper

@execution_timer
def process_loop():
    sum(range(1000000))

process_loop()

```

### Q18: What is the purpose of the `functools.wraps` decorator inside custom wrappers? What breaks if you omit it?

#### Answer:

When you wrap a function inside a decorator, the original function is overwritten by the inner wrapper function object.

If you omit `@wraps(func)`, the target function **loses its identity metadata**. Its `__name__` property becomes `"wrapper"` and its `__doc__` docstring documentation gets erased. This breaks reflection tools, debuggers, and automated documentation generators.

### Q19: What is the structural difference between `Callable[[int, int], int]` and a standard type hint label like `int`? Where is it imported from?

#### Answer:

* **`Callable`** is used to type hint **functions themselves** (objects that can be invoked with parentheses). The syntax `Callable[[int, int], int]` states: *"This variable must be a function that accepts two integers as arguments and returns an integer."*
* Standard labels like `int` simply reference static variable data instances.
* **Importing:** It is imported from the standard library's **`typing`** module (or accessed via `collections.abc.Callable` in modern Python versions).

### Q20: What will the following expression evaluate to, and what does it reveal about Boolean evaluations in Python?

```python
def check_status():
    print("Function Evaluated!")
    return True

result = False and check_status()

```

#### Answer:

* **Output:** `result` becomes `False`. Nothing is printed to the terminal screen.
* **Explanation:** This reveals **Short-Circuit Evaluation**. Because the left side of the `and` statement is already `False`, the entire expression can never evaluate to `True`. Python optimizes performance by skipping the execution of `check_status()` entirely.
