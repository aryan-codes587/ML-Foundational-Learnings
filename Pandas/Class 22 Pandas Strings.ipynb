```python
df = pd.read_csv('titanic.csv')
df.head(3)
```

```python
# common functions
# lower/upper/capitalize/title

df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
df['Name'].str.lower()

# len
# find the longest name
df.iloc[df['Name'].str.len().argmax()]['Name']
# df[df['Name'].str.len() == 82]['Name']

# strip
df['Name'].str.strip()
```

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0) # gets all 0th index item of all list
df.head()
```

```python
df[['title','firstname']] =  df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1 , expand=True) # n means -> how many split you want to do.
df.head(3)
```

```python
df['title'].value_counts()
```

```python
# replace
# Miss and Ms both are same
df['title']=df['title'].str.replace('Ms.','Miss.')
df['title']=df['title'].str.replace('Mlle.','Miss.')
df['title'].value_counts()
```

```python
# filtering
# startswith / endswith
df[df['firstname'].str.startswith('A')].head()
```

```python
df[df['firstname'].str.endswith('A')].head()
```

```python
df[df['firstname'].str.isdigit()].head()
```

```python
# applying regex
# contains
# search john -> both case
df[df['firstname'].str.contains('john',case=False)].head()
```

```python
# find last names with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].head()
```
