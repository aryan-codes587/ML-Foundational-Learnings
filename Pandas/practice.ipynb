## Concatenation (Stacking the Data)
- `concat()` is like putting dataframes together without merging rows.


```python
import pandas as pd

df1 = pd.DataFrame({
    'Name':['Aryan','Raj'],
    'Age':[20,21]
})

df2 = pd.DataFrame({
    'Name':['Priya','Neha'],
    'Age':[22,23]
})
```

```python
pd.concat([df1,df2])
```

### Reset Index

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
# Horizontal concatenation
pd.concat([df1,df2],axis=1)
```

```python
# Join
# works using index
df1 = pd.DataFrame({
    'Name' : ['Aryan','Raj']
} , index=[1,2])

df2 = pd.DataFrame({
    'Marks' : [90,95]
} , index=[1,2])
```

```python
df1.join(df2)
```

```python
df1 = pd.DataFrame({
    'id' : [1,2,3],
    'name' : ['A','B','C']
})
marks = pd.DataFrame({
    'id' : [1,2,3],
    'marks' : [90,80,95]
})
pd.merge(df1,marks,on='id')
```
