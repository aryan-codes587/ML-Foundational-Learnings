### Pandas Series
- A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
country = ['India', 'Pakistan', 'USA' , 'Nepal']
pd.Series(country)
```

```python
# integers
runs = [12,24,5,78,100]
runs_ser = pd.Series(runs)
print(runs_ser)
```

```python
# custom index
marks = [65,57,89,100]
subjects = ['maths', 'english', 'science', 'hindi']

pd.Series(marks , index = subjects)
```

```python
# setting a name
marks = pd.Series(marks, index=subjects , name = 'aryan marks')
print(marks)
```

### Series from dict

```python
marks = {
    'maths' : 67,
    'english' : 57,
    'science' : 89,
    'hindi' : 100
}
marks_series = pd.Series(marks, name='aryan marks')
marks_series
```

```python
subs.clip(100,200)
```

Here is a quick breakdown of what `subs.clip(100, 200)` is doing:

* **What it does:** It caps (or trims) the values in your Series to sit strictly within a specified floor (lower bound) and ceiling (upper bound).
* **The lower bound (`100`):** Any value in the Series that was *less than 100* is automatically bumped up to exactly `100` (which is why indices 0 through 4, which were originally 48, 57, 40, 43, and 44, all became `100`).
* **The upper bound (`200`):** Any value in the Series that was *greater than 200* is forced down to exactly `200` (which is why indices 360 and 361, originally 231 and 226, dropped down to `200`).
* **The middle values:** Any value already between 100 and 200 stays exactly the same (like indices 362, 363, and 364).

> **Note:** By default, `.clip()` returns a *copy* of the modified data. It doesn't change your original `subs` Series unless you assign it back (`subs = subs.clip(100, 200)`) or use the parameter `inplace=True`.
