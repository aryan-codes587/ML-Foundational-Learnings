### What is an API?

An API (Application Programming Interface) is a set of rules that allows different software applications to communicate with each other. It acts as a bridge between a client and a server, enabling one application to request data or services from another without needing to know how the underlying system works. For example, when a weather app shows current weather information, it usually gets that data from a weather API. APIs make software development faster, more secure, and more organized by providing a standard way for applications to exchange information.

```python
import requests
import pandas as pd

API_KEY = "7b9d91bc379f746baff9690bb55ef791"

# First request to get total pages
url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&language=en-US&page=1"

response = requests.get(url)
data = response.json()

total_pages = data['total_pages']

dfs = []

for page in range(1, total_pages + 1):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&language=en-US&page={page}"

    response = requests.get(url)
    data = response.json()
    temp_df = pd.DataFrame(data['results'])[
        [
            'id',
            'title',
            'overview',
            'release_date',
            'popularity',
            'vote_average',
            'vote_count'
        ]
    ]

    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)

print(df.shape)
print(df.head())
```
