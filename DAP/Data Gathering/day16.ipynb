## 2. Working with JSON :- 
* A lightweight format for **storing and exchanging data**.
* Human-readable and easy for machines to parse.
* Uses **key-value pairs**.
* Commonly used in **APIs**, web applications, and configuration files.
* Supports these data types:

  * String
  * Number
  * Boolean (`true`, `false`)
  * Array (`[]`)
  * Object (`{}`)
  * Null (`null`)

**Example:**

```json
{
  "name": "Aryan",
  "age": 20,
  "skills": ["Python", "JavaScript"],
  "isStudent": true
}
```

**Python Conversion:**

```python
import json

# Python dict → JSON
json_data = json.dumps(data)

# JSON → Python dict
python_data = json.loads(json_string)
```

**Remember:**
JSON is essentially a **text representation of a dictionary/object** used for data exchange between systems.
