```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Colored Scatter-plots

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['species'] = iris['species'].replace({'setosa' : 0, 'versicolor':1, 'virginica':2})
iris
```

```python
plt.scatter(iris['sepal_length'], iris['petal_length'], c=iris['species'], cmap='jet', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

#### Plot size

#### Annotation

```python
batters = pd.read_csv('batter copy.csv')
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

#### 3D surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
# print(xx[0][1])
```

```python
z =  xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

#### Contour Plots

```python
# Create the figure
fig = plt.figure(figsize=(12, 8))

# Add a subplot to the figure
ax = plt.subplot()

# Generate the contour plot
p = ax.contour(xx, yy, z, cmap="viridis")

# Add the colorbar directly to the figure, referencing the plot
fig.colorbar(p, ax=ax)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

z = xx**2 + yy**2

# Create the figure
fig = plt.figure(figsize=(12, 8))

# Add a subplot to the figure
ax = plt.subplot()

# Generate the contour plot
p = ax.contour(xx, yy, z, cmap="viridis")

# Add the colorbar directly to the figure, referencing the plot
fig.colorbar(p, ax=ax)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

#### Heatmap

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df.head(3)
```

```python
grid = temp_df.pivot_table(index = 'overs', columns='ballnumber', values='batsman_run', aggfunc='count')
grid.head(3)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6) , list(range(1,7)))
plt.colorbar()
```
