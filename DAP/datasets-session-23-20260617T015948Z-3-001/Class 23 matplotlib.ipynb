### Types of Data :- 
- Broadly there are two types of data :-
    - **Numerical Data** :- Continous or discrete numbers (e.g., prices, runs scored, age).

### Bar Charts (Categorical Comparisons)
- Bar Charts display data using rectangular bars where the length of each bar is proportional to the value it represents
    - **Analysis Type** :- Bivariate Analysis (Numerical VS Categorical)
    - **Primary Use** :- Comparing independent groups, ranks, or aggregate tools.

````python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red', 'blue', 'green' , 'yellow', 'pink']

plt.bar(colors, children, color='blue')
````

````python
# horizontal bar chart
plt.barh(colors,children,color='black')
````

````python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
````

````python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'] , width=0.2 , color='yellow', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'] , width=0.2 , color='red', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'] , width=0.2 , color='blue', label='2017')
plt.xticks(np.arange(df.shape[0]) , df['batsman'])
# plt.plot()
plt.legend()
plt.show()
````

This code snippet creates a **grouped (multi-series) bar chart** using Matplotlib. Instead of stacking the bars on top of each other, it displays the data for three different years (2015, 2016, and 2017) **side-by-side** for each batsman.

Here is a line-by-line breakdown of exactly how it works:

---

### 1. The Core Concept: Positioning with `np.arange()`

If you pass the same X-coordinates to all three `plt.bar()` functions, the bars will overlap and hide one another. To prevent this, the code uses a math trick to shift each year's bars slightly to the left, center, and right.

* **`df.shape[0]`**: This returns the number of rows (or unique batsmen) in your DataFrame. Looking at your dataset, this value is `5`.
* **`np.arange(df.shape[0])`**: This generates an array of indices for the X-axis: `[0, 1, 2, 3, 4]`. These act as temporary numeric anchors for each player.
* **`width = 0.2`**: This explicitly sets the thickness of every single bar to `0.2` units.

---

### 2. Positioning the Bars Side-by-Side

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='yellow')

```

* **What it does:** Shifts the X-coordinates for the 2015 data backwards by `0.2` (resulting in `[-0.2, 0.8, 1.8, 2.8, 3.8]`).
* **Result:** The yellow bars for 2015 are pushed to the **left** of the main anchor.

```python
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='red')

```

* **What it does:** Keeps the default X-coordinates exactly at `[0, 1, 2, 3, 4]`.
* **Result:** The red bars for 2016 sit perfectly in the **center** anchor point.

```python
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='blue')

```

* **What it does:** Shifts the X-coordinates for the 2017 data forward by `0.2` (resulting in `[0.2, 1.2, 2.2, 3.2, 4.2]`).
* **Result:** The blue bars for 2017 are pushed to the **right** of the main anchor.

Because each bar has a width of exactly `0.2`, they line up perfectly next to each other without gaps or overlapping edge-to-edge ($0.2 + 0.2 + 0.2 = 0.6$ total width per group).

---

### 3. Fixing the X-Axis Labels

```python
plt.xticks(np.arange(df.shape[0]), df['batsman'])

```

* If you skip this line, your X-axis ticks will just show the raw placeholder numbers: `0, 1, 2, 3, 4`.
* `plt.xticks()` tells Matplotlib: *"Go to positions `[0, 1, 2, 3, 4]` (the exact center points where the 2016 red bars sit) and swap out those numbers for the actual textual names in the `df['batsman']` column (e.g., AB de Villiers, V Kohli)."*

---

### 4. Rendering the Visual

```python
plt.show()

```

* This clears the background buffer and explicitly draws the finished coordinate canvas onto your screen in the Colab notebook.
