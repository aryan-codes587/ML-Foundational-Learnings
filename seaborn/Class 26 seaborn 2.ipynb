```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
tips
```

### Categorical Plots

#### Categorical Scatter Plot
- Stripplot
- Swarmplot

#### Categorical Distribution Plots
- Boxplot
- violinplot

#### Categorical Estimate Plot -> for central tendancy
- Barplot
- Pointplot
- Countplot

#### Figure level function -> `catplot`

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# strip plot :- Plots scatter plot between a numrical VS categorical column
# axes level function
sns.stripplot(data=tips , x='day', y='total_bill')
```

```python
# using catplot
# figure level function
sns.catplot(data = tips, x='day', y='total_bill', kind='strip')
```

```python
# jitter 
sns.catplot(data = tips, x='day', y='total_bill', kind='strip',jitter=0.2,hue='sex')
```

```python
# swarmplot
sns.catplot(data = tips, x='day', y='total_bill', kind='swarm')
```

```python
# hue
sns.swarmplot(data = tips, x='day', y='total_bill',hue='sex')
```

### Box plot:-
- A box plot is a standardized way of displaying the distribution of data based on a five number summary ('minimum', first quartile [Q1], median, third quartile [Q3], and 'maximum'). 
- It can tell you about your outliers and what their values are. Box plots can also tell you if your data is symmetrical, how tightly you data is grouped and if and how your data is skewed.



```python
# Box plot
sns.boxplot(data = tips, x='day', y='total_bill')
```

```python
# using cat plot
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
# hue
sns.boxplot(data = tips, x='day', y='total_bill',hue='sex')
```

```python
# single box plot -> numerical columns
sns.boxplot(data = tips, y='total_bill')
```

#### Violin Plot = (Boxplot + KDE)

```python
sns.set_palette('tab10')
```

```python
# violin plot
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data = tips, x='day', y='total_bill', kind='violin')
```

```python
# hue 
sns.catplot(data = tips, x='day', y='total_bill',kind='violin',hue='sex',split=True,col='day',col_wrap=2)
```

```python
# bar plot
import numpy as np
sns.barplot(data = tips, x='sex', y='total_bill', hue='smoker',estimator=np.mean) # bars are std of the distribution
```

```python
# point plot
sns.pointplot(data = tips, x='sex', y='total_bill', hue='smoker',estimator=np.mean)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# countplot
sns.countplot(data = tips, x='sex',hue='day')
```

- A special case for the bar plot is when you want to show the number of observations in each category rather tahn computing a statistic for a second variable. 
- This is similar to a histogram over a categorical, rather than quantitative variable.

```python
# faceting using catplot
sns.catplot(data = tips, x='sex', y='total_bill', col='smoker',row='time',kind='box',hue='sex')
```

### Regression plots :-
- regplot
- Implot
- In the simplest innovation, both function draw a scatter plot of two variables, x and y and then fit the regerssion model y ~ x and plot the resulting regression line and a 95% confidence interval for the regression.

```python
# axes level 
# hue parameter is not available
sns.regplot(data = tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data = tips, x='total_bill', y='tip', hue='sex')
```

```python
# residplot
sns.residplot(data = tips, x='total_bill', y='tip')
```

**A second way to plot Facet Plots -> FacetGrid**

```python
# figure level -> relplot -> displit -> catplot -> lmplot
```

```python
sns.catplot(data = tips, x='sex', y='total_bill', kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data = tips, col='day', row='time', hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
```

#### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data = iris, hue='species')
g.map(sns.scatterplot)
g.add_legend()
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

#### JointGrid VS JointPlot

```python
sns.jointplot(data = tips, x='total_bill', y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data = tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.violinplot)
```
